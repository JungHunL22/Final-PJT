```python
def outlier(method=None,data,col):
    if method=='IQR':
        Q1 = np.percentile(data[col], 25)
        Q3 = np.percentile(data[col], 75)
        IQR = Q3 - Q1

        out=data[(data[col]<=(Q1-1.5*IQR))|(data[col]>=(Q3+1.5*IQR))]
    
    return out
```
